# Scatter Plot
Scatter plots are useful for visualizing 'clusters' of data. They map out examples from a dataset on an (x, y) coordinate plane.
These are commonly used for visualizing commonalities or differences in data, such as age groups in a customer base. They have plenty of other uses and are commonly known as a versatile plot. 

## Handle Imports
First we have to handle our dependencies. See below. 

```python
import pandas as pd # naming convention
from matplotlib import pyplot as plt # another naming convention
```

### Now we will set up a sample data set

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

### Set up labels and title
Use pyplot object oriented programming to create labels for the plot. 

```python
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

# Tighten up the plot layout and create it as a scatter plot
plt.tight_layout()
plt.scatter(x, y, s=100, c='green')
```
